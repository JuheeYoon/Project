# Airline Passenger Satisfaction 데이터 설명

  - 게이트 위치(Gate location) : 만족도
  - 식사, 음료(Food and drink) : 만족도
  - 온라인 탑승(Online boarding) : 만족도
  - 좌석의 편안함(Seat comfort) : 만족도
  - 기내 엔터테인먼트(Inflight entertainment) : 만족도
  - 탑승 서비스(On-board service) : 만족도
  - 다리를 뻗을 공간(Leg room service) : 만족도
  - 수화물 처리(Baggage handling) : 만족도

  - 기내 서비스(Inflight service) : 만족도
  - 청결함(Cleanliness) : 만족도
  - 출발 지연 시간(Departure Delay in Minutes) : 분 단위
  - 도착 지연 시간(Arrival Delay in Minutes) : 분 단위
  - 만족도 (Satisfaction) : 만족(satisfied), 중립(neutral) & 불만족(dissatisfaction)

# **2. 데이터 정리**

1. Unnamed: 0 열 & id 열삭제
2. Arrival Delay in Minutes에 있는 Null 값 행을 제거 !

3. Gender 레이블 인코딩
4. Customer Type 레이블 인코딩
5. Type of Travel 레이블 인코딩
6. Class 레이블 인코딩
7. satisfaction 레이블 인코딩

8. Age : UN 나이 분류에 따름 ⇒ 0∼17세 = 미성년자(0), 18∼65세 = 청년(1), 66∼79세 = 중년(2), 80세∼ = 노년(3)

9. Flight Distance : 0 ~ 1 값으로 정규화
10. Departure Delay in Minutes : 0 ~ 1 값으로 정규화
11. Arrival Delay in Minutes : 0 ~ 1 값으로 정규화

```python
# 1. Unnamed: 0 열 & id 열 삭제
data = passenger.drop(['Unnamed: 0','id'], axis = 1)

# 2. 결측치 포함 행 제거
data.dropna()

### 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

# 3. Gender 레이블 인코딩
encoderG = LabelEncoder()
encoderG.fit(data['Gender'])
data['Gender'] = encoderG.transform(data['Gender'])
print(encoderG.classes_)

# 4. Customer Type 레이블 인코딩
encoderCT = LabelEncoder()
encoderCT.fit(data['Customer Type'])
data['Customer Type'] = encoderCT.transform(data['Customer Type'])
print(encoderCT.classes_)

# 5. Type of Travel 레이블 인코딩
encoderTT = LabelEncoder()
encoderTT.fit(data['Type of Travel'])
data['Type of Travel'] = encoderTT.transform(data['Type of Travel'])
print(encoderTT.classes_)

# 6. Class 레이블 인코딩
encoderC = LabelEncoder()
encoderC.fit(data['Class'])
data['Class'] = encoderC.transform(data['Class'])
print(encoderC.classes_)
#ClassMap = {"Business" : 0, "Eco" : 1, "Eco Plus" : 2}
#data['Class'] = data['Class'].map(ClassMap)

# 7. satisfaction 레이블 인코딩
encoderS = LabelEncoder()
encoderS.fit(data['satisfaction'])
data['satisfaction'] = encoderS.transform(data['satisfaction'])
print(encoderS.classes_)

# 8. Age 매핑
def get_category(age):
  cat = int()
  if age <= 17: cat = 0
  elif age <= 65 : cat = 1
  elif age <= 79 : cat = 2
  else : cat = 3
  return cat

data['Age'] = data['Age'].apply(lambda x : get_category(x))

### 정규화 스케일링
# 9. Flight Distance 스케일링
data['Flight Distance'] = (data['Flight Distance'] - data['Flight Distance'].min()) / (data['Flight Distance'].max() - data['Flight Distance'].min())

# 10. Departure Delay in Minutes 스케일링
data['Departure Delay in Minutes'] = (data['Departure Delay in Minutes'] - data['Departure Delay in Minutes'].min()) / (data['Departure Delay in Minutes'].max() - data['Departure Delay in Minutes'].min())

# 11. Arrival Delay in Minutes 스케일링
data['Arrival Delay in Minutes'] = (data['Arrival Delay in Minutes'] - data['Arrival Delay in Minutes'].min()) / (data['Arrival Delay in Minutes'].max() - data['Arrival Delay in Minutes'].min())

### 확인 !
data
```
